```python
#Prime Number
def is_prime(num):
    if (num<=1):
        return False
    
    for i in range(2,num):
        if (num % i == 0):
            return False
    return True
    
num = int(input("Enter Number : "))
if is_prime(num):
    print("The number {} is Prime".format(num))
else:
    print("The number {} is not Prime".format(num))
```
